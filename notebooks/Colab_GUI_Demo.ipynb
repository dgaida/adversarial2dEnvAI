## 2. Launch the Interactive GUI

The `ColabGUI` class provides an easy way to visualize and interact with the environment. It includes buttons for stepping through the environment, resetting the episode, and toggling the Particle Filter visualization.

```python
from custom_grid_env.colab_gui import ColabGUI
from custom_grid_env.agents.random_player_agent import RandomPlayerAgent

# Initialize and run the GUI
gui = ColabGUI(agent_class=RandomPlayerAgent, slip_probability=0.1)
gui.run()
```

## 3. GUI Features

- **Next Step**: Executes one action from the agent and updates the visualization.  
- **Reset Episode**: Resets the environment to its initial state.  
- **Show Particles**: Toggles the display of particles from the Particle Filter.  
- **PF Sensors**: Choose which sensors the Particle Filter should use:  
    - **Neural Net**: Uses CNN predictions for localization.  
    - **Color Sensor**: Uses ground color sensor for localization.  
    - **Both**: Uses both sensors for improved localization accuracy.  
- **Agent Knowledge**: Choose whether the agent should use the actual environment state or its own estimate (from the Particle Filter) for decision making.
    - **Actual State**: Agent uses the true ground-truth position.
    - **Estimated State**: Agent uses its estimated position and the ghost position relative to that estimate.
- **Stats**:  Displays the current step count and accumulated reward.  
- **Deterministic Mode**: Movement is 100% accurate, no slipping.  
- **No Ghost Mode**: The ghost is removed, turning it into a single-agent environment.  
- **Task Planning**: Enter a natural language task like "Besuche das Feld mit der Klaviermusik und dann das Ziel". An LLM will identify the targets, and a TSP solver + Value Iteration will execute the optimal path.
