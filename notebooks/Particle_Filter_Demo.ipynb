## 3. Features & Controls

### Navigation & Execution
- **Next Step**: Executes one action from the agent and updates the visualization.
- **Execute**: Automatically navigates the agent to the manually set goal.
- **Reset Episode**: Resets the environment and the particle filter.
- **Manual Goal**: Enter the row and column coordinates for the target and click **Set Goal**.

### Particle Filter Settings
- **Show Particles**: Toggles the display of individual particles on the grid.
- **Particles**: Choose the number of particles (20, 40, 100, 200). A higher number generally leads to better localization but requires more computation.
- **PF Sensors**: Choose which sensors the Particle Filter should use:
    - **Neural Net**: Uses CNN predictions for localization.
    - **Color Sensor**: Uses ground color sensor for localization.
    - **Both**: Uses both sensors for improved localization accuracy.
- **Color Acc**: Adjust the accuracy of the color sensor (from 70% to 100%).

### Movement & Uncertainty
- **Deterministic Mode**: If checked, movement is 100% accurate (no slipping).
- **Slip Type**: Choose between **Longitudinal** (moving twice or staying in place) and **Perpendicular** (moving sideways) slipping models.
- **Agent Knowledge**: 
    - **Actual State**: Agent uses the true position for planning.
    - **Estimated State**: Agent uses the position estimated by the Particle Filter.
