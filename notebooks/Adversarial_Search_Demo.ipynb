## 2. Launch the Adversarial Search GUI

The `AdversarialColabGUI` class provides an interface for exploring how adversarial agents perform under uncertainty. You can switch between Minimax and Expectimax agents, configure the ghost's behavior, and decide whether the agent uses the ground truth or the Particle Filter's estimated position for decision making.

```python
from custom_grid_env.colab_gui import AdversarialColabGUI
from custom_grid_env.agents.adversarial_agents import MinimaxAgent

# Initialize and run the Adversarial Search GUI
gui = AdversarialColabGUI(agent_class=MinimaxAgent, slip_probability=0.1)
gui.run()
```

## 3. Features & Controls

### Navigation & Execution
- **Next Step**: Executes one action from the agent and updates the visualization.
- **Execute**: Automatically navigates the agent to the manually set goal while avoiding the ghost.
- **Reset Episode**: Resets the environment, the particle filter, and the ghost position.
- **Manual Goal**: Enter the row and column coordinates for the target and click **Set Goal**.

### Agent & Ghost Settings
- **Agent Type**: Choose between **Minimax**, **Expectimax**, and other agents.
- **Ghost Type**: Configure the ghost behavior (**Chase**, **Random**, or **Minimax**).
- **Use Ghost**: Toggle whether the ghost is active in the environment.
- **Agent Knowledge**: 
    - **Actual State**: Agent uses the true position for planning.
    - **Estimated State**: Agent uses the position estimated by the Particle Filter.

### Particle Filter & Uncertainty
- **Show Particles**: Toggles the display of individual particles on the grid.
- **Particles**: Choose the number of particles (20, 40, 100, 200).
- **PF Sensors**: Choose which sensors (Neural Net, Color Sensor, or both) the Particle Filter should use.
- **Color Acc**: Adjust the accuracy of the color sensor.
- **Deterministic Mode**: If checked, movement is 100% accurate (no slipping).
- **Slip Type**: Choose between **Longitudinal** and **Perpendicular** slipping models.
